```python
import pandas as pd
```

```python
#Extraer la tabla json con la metadata.
df = pd.read_json(path_or_buf='https://api.stackexchange.com/2.2/search?order=desc&sort=activity&intitle=perl&site=stackoverflow', orient='records')

#Convertir la primera fila a lista
lista = list(df['items'])

#Convertir la lista de diccionarios a dataframe.
df_lista = pd.json_normalize(lista)

#Convertir las fechas a formato timestamp.
df_lista['last_activity_date'] = pd.to_datetime(df_lista['last_activity_date'], unit='s')
df_lista['creation_date'] = pd.to_datetime(df_lista['creation_date'], unit='s')

df_lista
```

```python
#Obtener el número de respuestas contestadas y no contestadas.
    #Respuestas contestadas
df_preguntas_contestadas = df_lista.loc[df_lista['is_answered'] == True]
preguntas_contestadas = df_preguntas_contestadas['is_answered'].count()
    #Respuesta
df_preguntas_nocontestadas = df_lista.loc[df_lista['is_answered'] == False]
preguntas_nocontestadas = df_preguntas_nocontestadas['is_answered'].count()
```

```python
#Obtener la respuesta con menor numero de vistas
df_menos_vistas = df_lista.sort_values(by='view_count')
df_menos_vistas = df_menos_vistas.head(1)
df_menos_vistas = df_menos_vistas['question_id']
menos_vistas = df_menos_vistas.to_string(index = False)
```

```python
#Obtener la respuesta mas vieja y la mas actual

    #respuesta mas vieja
df_respuesta_vieja = df_lista.sort_values(by='creation_date')
df_respuesta_vieja = df_respuesta_vieja.head(1)
df_respuesta_vieja = df_respuesta_vieja['question_id']
respuesta_vieja = df_respuesta_vieja.to_string(index = False)

    #respuesta mas nueva
df_respuesta_actual = df_lista.sort_values(by='creation_date', ascending = False)
df_respuesta_actual = df_respuesta_actual.head(1)
df_respuesta_actual = df_respuesta_actual['question_id']
respuesta_actual = df_respuesta_actual.to_string(index = False)
```

```python
# Obtener la respuesta del owner que tenga mayor reputacion:
df_mas_reputacion = df_lista.sort_values(by='owner.reputation', ascending = False)
df_mas_reputacion = df_mas_reputacion.head(1)

df_owner_mas_reputacion = df_mas_reputacion['owner.user_id']
df_mas_reputacion = df_mas_reputacion['question_id']

mas_reputacion = df_mas_reputacion.to_string(index = False)
owner_mas_reputacion = df_owner_mas_reputacion.to_string(index = False)
```

```python
print('1.- Obtener la respuesta con menor numero de vistas: ')
print('Preguntas Contestadas: ' + str(preguntas_contestadas) +'  ||  Preguntas No Contestadas: ' + str(preguntas_nocontestadas))
print(' ')
print('2.- Obtener la respuesta con menor número de vistas: ')
print('Question id de la respuesta con menor número de vistas: ' + str(menos_vistas))
print( ' ')
print('3.- Obtener la respuesta mas vieja y la mas actual: ')
print('Question id de la respuesta más vieja: ' +  str(respuesta_vieja) + '  ||  Question id de la respuesta más nueva: ' + str(respuesta_actual))
print( ' ')
print('4.- Obtener la respuesta del owner que tenga mayor reputación: ')
print('Question id de la respuesta del owner con mas reputacion: ' + str(mas_reputacion) + '  ||  Owner_id: ' + str(owner_mas_reputacion))
```
